```python
from keras.models import Sequential
from keras.layers import Dense, Activation
from sklearn.metrics import confusion_matrix
from sklearn.metrics import f1_score
from sklearn.metrics import accuracy_score
from sklearn.model_selection import train_test_split
import numpy as np
import pandas as pd
dataset = pd.read_excel("/content/Data_BM.csv",header = 0, sheet_name = "Sheet1")
dataset.head()
```

```python
model = Sequential()
model.add(Dense(20, input_shape=(16,)))
model.add(Activation('sigmoid'))
model.add(Dense(1))
model.add(Activation('sigmoid')) 

# Compile the model and calculate its accuracy:
model.compile(loss='mean_squared_error', optimizer='sgd', metrics=['accuracy']) 

# Print a summary of the Keras model:
model.summary()
```

```python
x = np.array(dataset[["x1","x2","x3","x4","x5","x6","x7","x8","x9","x10","x11","x12","x13","x14","x15","x16"]])
y = np.array(dataset.Classlabel);
```

```python
x_train, x_test, y_train, y_test= train_test_split(x, y, train_size=0.8)
```

```python
print(len(x_train))
print(len(y_train))
print(len(x_test))
print(len(y_test))
```

```python
mdl = model.fit(x_train,y_train, epochs=150)
```

```python
y_pred = model.predict(x_test)
```

```python
y_pred
```

```python
y_pclass = np.where(y_pred>=0.5,1,0)
```

```python
cm= confusion_matrix(y_test,y_pclass)
print (cm)
print(f1_score(y_test, y_pclass))
```

```python
from keras.metrics.metrics import accuracy
print(accuracy_score(y_test, y_pclass))
```

```python
from sklearn.metrics import classification_report
```

```python
print(classification_report(y_test,y_pclass))
```
